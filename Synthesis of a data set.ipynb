In this notebook I am going to look at a real phenomenon of my choice: a health study with a sample of population related to heart disease. A heart attack is usually caused by the buildup of plaque in the coronary arteries. The plaque builds up over time to eventually slow or completely block blood flow to the heart muscle. What matters now is the variables that condition the appearance of this health issue. It's worth mentioning that there may not be a most definitive one by itself, it seems to be just a question of balance between themselves. Anyways they are approximately graded from greater to lesser likelihood of importance.

For the purpose of this exercise I will synthesise a data set with the values of all the risk factor which affect the heart attack. Half of the samples represent males and the other half woman being 400 in total.

## 1) Age
As we get older, our risk of a heart attack goes up. Men age 45 or older and women age 55 or older are more likely to have a heart attack than are younger men and women [[1]](https://www.mayoclinic.org/diseases-conditions/heart-attack/symptoms-causes/syc-20373106) This is a discrete numerical variable and for the synthetised data set I am going to generate a sample of 400 people. The age will be presented in years with integer values and following a uniform distribution to garantee that all the age ranges will be equally represented.

```python
import numpy as np
age = np.random.randint(20,70,size=400)
age
```

```python
import matplotlib.pyplot as plt
plt.hist(age)
plt.show()
```

## 2) Gender
Men face a greater risk of heart disease than women. On average, a first heart attack strikes men at age 65. For women, the average age of a first heart attack is 72 [[2]](https://www.health.harvard.edu/heart-health/the-heart-attack-gender-gap). This is a non-numerical parameter -a string- which follows a Bernoulli distribution since there are just 2 possible outcomes (man and woman) being one of them more likely to occur (men).

```python
possible_genders = ['male','female']
gender =  [np.random.choice(possible_genders) for i in range(400)]
gender
```

```python
import seaborn as sns
import pandas as pd
df = pd.DataFrame({"gender": gender})

# The creation of the plot takes just two lines
sns.set_style("whitegrid")
sns.countplot(x="gender", data=df)
plt.show()
```

## 3) Family history
Having a family history of cardiovascular disease increases the risk of developing cardiovascular diseases such as heart attacks. Genes can pass on the risk of cardiovascular disease, and they can also be responsible for passing on other conditions such as high blood pressure or high cholesterol levels [[3]](https://www.bhf.org.uk/informationsupport/risk-factors/family-history). 

For this excercise I will consider just 2 possible outcomes, the existance of a family history as 'Yes' or its absence as 'No'. This means that this variable holds strings however the likelyhood of both outcomes are not balanced. I could try to roughly estimate this by considering some facts from the Irish population like:

- Around 10.000 people die every year for cardiovascular disorders [[4]](https://www.hse.ie/eng/health/az/c/coronary-heart-disease/)
- The population in 2016 was nearly 4.800.000 according to CSO [[5]](https://www.cso.ie/en/releasesandpublications/ep/p-cpr/censusofpopulation2016-preliminaryresults/)

But that would lead to a different project so for the puropose of this test the following figure doesn't need to be representative, it just means the difference in probability of every ocurrence. Then let's just say that a 30% of the people in Ireland has relatives who have had cardiovascular issues including all ranges of severity. Then the likelyhood of having a family history is 30%, and therefore the probability of not having a family history is 70%

```python
possible_fHist = ['Yes','No']
fHist = np.random.choice(possible_fHist, 400, p=[0.3, 0.7])
df = pd.DataFrame({"Family history": fHist})
sns.countplot(x="Family history", data=df)
plt.show()
```

## 4) Blood Cholesterol

LDL cholesterol travels from the liver through the arteries to other parts of our body. LDL is called bad cholesterol because it sticks to the walls in our arteries making them narrow. The image on the right shows the ranges of blood LDL cholesterol:
 
In order to force the data set to be more representative of reality I am going to align this variable with the other two variables *age* and *gender* according to the statement given at the top of this document in the "Age" section: *Men age 45 or older and women age 55 or older are more likely to have a heart attack than are younger men and women* [[1]](https://www.mayoclinic.org/diseases-conditions/heart-attack/symptoms-causes/syc-20373106). This way I will assign higher  levels of this bad cholesterol to men older than 44 and women older than 54 because they are more likely to have a heart attack. 

When it comes to this variable itself, it is numerical and depending on how specific the machine that processes the blood sample is, we could label it as discrete or continuous. This is, if it returns an integer, then it would be discrete but if the figure has several decimals, then we can take it as continuous.

```python
bloodChol = np.zeros([400])
# Create a table with the synthetised data set
df = pd.DataFrame({'age': age, 'gender': gender, 'Family history': fHist, 'Blood cholesterol': bloodChol})
# Align the columns at the center with the style feature
df.style.set_properties(**{'text-align': 'center'})
```

## 5) Smoking

The results of Healthy Ireland survey published in 2015 found that 19% of the Irish population (aged 15 and over) smoke on a daily basis. A further 4% identify themselves as occasional smokers [[6]](https://health.gov.ie/healthy-ireland/tobacco/smoking-prevalence-rates/). Then, regardless of the number of cigarettes they consum I will consider the consumption frequency only, being therefore a variable that holds strings instead of numerical values. 

For this I'll assign each of the three possible outcomes a different probability according to the result of the Healthy Ireland survey quoted above:

```python
# First I declare three variables that represent each consumption frequency
consumption_type = ['Regular','Occasional','Non-smokers']

# Generation of the array that contains the 400 samples with their corresponding probability
smokers = np.random.choice(consumption_type, 400, p=[0.19, 0.04, 0.77])
df = pd.DataFrame({"Tobacco consumption": smokers})
sns.countplot(x="Tobacco consumption", data=df)
plt.show()
```
