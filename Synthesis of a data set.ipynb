In this notebook I am going to look at a real phenomenon of my choice: a health study with a sample of population related to heart disease. A heart attack is usually caused by the buildup of plaque in the coronary arteries. The plaque builds up over time to eventually slow or completely block blood flow to the heart muscle. What matters now is the variables that condition the appearance of this health issue. It's worth mentioning that there may not be a most definitive one by itself, it seems to be just a question of balance between themselves. Anyways they are approximately graded from greater to lesser likelihood of importance.

For the purpose of this exercise I will synthesise a data set with the values of all the risk factor which affect the heart attack. Half of the samples represent males and the other half woman being 400 in total.

## 1) Age
As we get older, our risk of a heart attack goes up. Men age 45 or older and women age 55 or older are more likely to have a heart attack than are younger men and women [[1]](https://www.mayoclinic.org/diseases-conditions/heart-attack/symptoms-causes/syc-20373106) This is a discrete numerical variable and for the synthetised data set I am going to generate a sample of 400 people. The age will be presented in years with integer values and following a uniform distribution to garantee that all the age ranges will be equally represented. And the numpy `random.randint` function provides me with this.

```python
# First of all I will import all the libraries I will use in this file
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# python magic command to make sure all the plots are displayed properly:
%matplotlib inline
# Create a 400 people age between 20 and 70 years old.
age = np.random.randint(20,70,size=400)
age
```

```python
plt.hist(age)
plt.show()
# we check visually that this numpy random function generates a uniform distribution.
```

## 2) Gender
Men face a greater risk of heart disease than women. On average, a first heart attack strikes men at age 65. For women, the average age of a first heart attack is 72 [[2]](https://www.health.harvard.edu/heart-health/the-heart-attack-gender-gap). This is a non-numerical parameter -a string- which follows a uniform distribution again, since for these kind of studies a similar number of group observations are preferred to make the comparison easier.

```python
# First declare the two genders
possible_genders = ['male','female']
# and then the corresponding array
gender =  [np.random.choice(possible_genders) for i in range(400)]
gender
```

```python
# Generation of a pandas data frame
df = pd.DataFrame({"gender": gender})

# The creation of the plot takes just two lines
sns.set_style("whitegrid")
sns.countplot(x="gender", data=df)
plt.show()
```

## 3) Family history
Having a family history of cardiovascular disease increases the risk of developing cardiovascular diseases such as heart attacks. Genes can pass on the risk of cardiovascular disease, and they can also be responsible for passing on other conditions such as high blood pressure or high cholesterol levels [[3]](https://www.bhf.org.uk/informationsupport/risk-factors/family-history). 

For this excercise I will consider just 2 possible outcomes, the existance of a family history as 'Yes' or its absence as 'No'. This means that this variable holds strings and the estimation of its distribution gets really challenging. I could try to roughly estimate this by considering some facts from the Irish population like:

- Around 10.000 people die every year for cardiovascular disorders [[4]](https://www.hse.ie/eng/health/az/c/coronary-heart-disease/)
- The population in 2016 was nearly 4.800.000 according to CSO [[5]](https://www.cso.ie/en/releasesandpublications/ep/p-cpr/censusofpopulation2016-preliminaryresults/)

But that would lead to a different project so for the puropose of this test the following figure doesn't need to be representative, it just means the difference in probability of every ocurrence. Then let's just say that a 30% of the people in Ireland has relatives who have had cardiovascular issues including all ranges of severity. Then the likelyhood of having a family history is 30%, and therefore the probability of not having a family history is 70%

```python
# array generation
possible_fHist = ['Yes','No']
fHist = np.random.choice(possible_fHist, 400, p=[0.3, 0.7])
# plot
df = pd.DataFrame({"Family history": fHist})
sns.countplot(x="Family history", data=df)
plt.show()
```

## 4) Blood Cholesterol

LDL cholesterol travels from the liver through the arteries to other parts of our body. LDL is called bad cholesterol because it sticks to the walls in our arteries making them narrow. The image on the right shows the ranges of blood LDL cholesterol:
 
In order to force the data set to be more representative of reality I am going to align this variable with the other two variables *age* and *gender* according to the statement given at the top of this document in the "Age" section: *Men age 45 or older and women age 55 or older are more likely to have a heart attack than are younger men and women* [[1]](https://www.mayoclinic.org/diseases-conditions/heart-attack/symptoms-causes/syc-20373106). This way I will assign higher  levels of this bad cholesterol to men older than 44 and women older than 54 because they are more likely to have a heart attack. 

When it comes to this variable itself, it is numerical and depending on how specific the machine that processes the blood sample is, we could label it as discrete or continuous. This is, if it returns an integer, then it would be discrete but if the figure has several decimals, then we can take it as continuous.

Then, before anything else I will create the data set table I have so far because only after this, I will be able to iterate row by row and create a new array depending on the values of the cells, and which will turn into the following column of the data.

```python
# Create a table with the synthetised data set
df = pd.DataFrame({'age': age, 'gender': gender, 'Family history': fHist})
# to start the index from 1 rather than 0
df.index = np.arange(1, len(df) + 1)
# Align the columns at the center with the style feature
df.style.set_properties(**{'text-align': 'center'})
```

Now I can proceed with the new array creation:

```python
# Synthesis of an empty array which will be populated by iterating the rows of the table above. 
bloodChol = []

# this function will generate the values and will be called every row with different arguments 
# depending on the gender and age
def choiceChol(a, b):
    x = np.random.randint(a,b) + np.random.rand()
    bloodChol.append(x)

for index, row in df.iterrows():
    if row.gender == 'male' and row.age > 45:
        choiceChol(159, 200)  
    elif row.gender == 'male' and row.age < 45:
        choiceChol(75, 150)
    elif row.gender == 'female' and row.age > 55:
        choiceChol(159, 200)
    else:
        choiceChol(75, 150)
        
bloodChol
```

```python
plt.hist(bloodChol, bins = 100)
df = pd.DataFrame({"Blood cholesterol": bloodChol})
plt.show()
```

There is a gap visible, I didn't assign any values between 149 and 158 for being the top of the borderline high level

## 5) Smoking

The results of Healthy Ireland survey published in 2015 found that 19% of the Irish population (aged 15 and over) smoke on a daily basis. A further 4% identify themselves as occasional smokers [[6]](https://health.gov.ie/healthy-ireland/tobacco/smoking-prevalence-rates/). Then, regardless of the number of cigarettes they consum I will consider the consumption frequency only, being therefore a variable that holds strings instead of numerical values. 

For this I'll assign each of the three possible outcomes a different probability according to the result of the Healthy Ireland survey quoted above:

```python
# First I declare three variables that represent each consumption frequency
consumption_type = ['Regular','Occasional','Non-smoker']

# Generation of the array that contains the 400 samples with their corresponding probability
smokers = np.random.choice(consumption_type, 400, p=[0.19, 0.04, 0.77])
df = pd.DataFrame({"Tobacco consumption": smokers})
sns.countplot(x="Tobacco consumption", data=df)
plt.show()
```

And the updated table with the last variable included is:

```python
df = pd.DataFrame({'age': age, 'gender': gender, 'Family history': fHist, 'Blood cholesterol': bloodChol, 'Tobacco consumption': smokers})
# to start the index from 1 rather than 0
df.index = np.arange(1, len(df) + 1)
# Align the columns at the center with the style feature
df.style.set_properties(**{'text-align': 'center'})
```

## 6) Obesity
Obesity is a risk factor for fatal heart attacks even for people who do not have the conditions normally associated with cardiovascular disease. A recent study published in April 2018 in the journal JAMA Cardiology [[7]](https://www.ncbi.nlm.nih.gov/pubmed/29490333) concluded that adults between ages 40 and 59 who are overweight or obese have a significantly increased risk (ranging from 21 to 85 percent higher) of developing cardiovascular disease as compared with their normal weight peers. Individuals who are overweight, defined as having a body mass index (BMI) between 25 and 29.9, or are obese (a BMI of 30 or higher), also have a much greater risk of developing cardiovascular disease at a younger age. 

Then I will generate the BMI for the sample of population considering the obesity prevalence of the Irish population according to this study shown in the image below [[8]](https://www.safefood.eu/SafeFood/media/SafeFoodLibrary/Documents/Professional/Nutrition/Adult-and-children-obesity-trends-ROI.pdf)

We see that the men data follows a triangular distribution with its peak on the right where the age range 50 - 59. However women's distribution tends to be more linear despite the age range 30 - 39 is slighty lower.

Then I'll apply the results of this study to give person by person a BMI depending on their age:

```python
'''Create an empty array whose values will be added by iteration with the help of a function. Note that
this variable has to be created out of the below function because it has to be visible for the rest of 
the code (global scope)'''
BMI = []

'''This function will be called one table row at a time to:
1) Decide whether that person will have a 'normal' BMI or an 'obese' one, given the probability of suffering
obesity considering gender and age
2) Generate a random number from the chosen BMI range'''

def choiceBMI(a, b):
    
    # Creation of labels for both BMI options
    altBMI = ['normal', 'obese']
    
    # Election itself (coin toss with a probability which will be assigned when calling this function)
    z = np.random.choice(altBMI, 1, p=[a, b])
    
    if z == 'normal':
        # The BMI is generated as the addition of an integer and a one-decimal float
        z2 = np.random.randint(20,30) + np.random.randint(10)/10
        BMI.append(z2)
        
    else:
        z2 = np.random.randint(30,36) + np.random.randint(10)/10
        BMI.append(z2)
    return z2;

# Iteration through the table row by row
for index, row in df.iterrows():
    
    if row.gender == 'male':
        if 20 < row.age < 29:
            # we call the above function considering the corresponding probabilities for gender and age
            choiceBMI(0.88,0.12)
        elif 30 < row.age < 39:
            choiceBMI(0.72,0.28)
        elif 40 < row.age < 49:
            choiceBMI(0.70,0.30)
        elif 50 < row.age < 59:
            choiceBMI(0.59,0.41)
        else:
            choiceBMI(0.69,0.31)
        
        
    else:
        if 20 < row.age < 29:
            choiceBMI(0.86,0.14)
        elif 30 < row.age < 39:
            choiceBMI(0.87,0.13)
        elif 40 < row.age < 49:
            choiceBMI(0.75,0.25)
        elif 50 < row.age < 59:
            choiceBMI(0.72,0.28)
        else:
            choiceBMI(0.77,0.23)
BMI
```

```python
df = pd.DataFrame({'age': age, 'gender': gender, 'Family history': fHist, 'Blood_cholesterol': bloodChol, 'Tobacco consumption': smokers, 'BMI': BMI})
# to start the index from 1 rather than 0
df.index = np.arange(1, len(df) + 1)
# Align the columns at the center with the style feature
df.style.set_properties(**{'text-align': 'center'})
```

And finally, while having a look at the table I spot some row values that could clash at first: mid-age people with high BMI but low blood cholesterol levels. In order to favour the known relationship of these risk factor with the discussed disease I will reset the values of the blood cholesterol cells from the mid-age people who have a high BMI but a low blood cholesterol, which is exactly what says the study [[7]](https://www.ncbi.nlm.nih.gov/pubmed/29490333) quoted above in point 6) Obesity. Then the following script will only modify certain values, remaining the others the same:

```python
newBloodChol = []
for index, row in df.iterrows():
    if row.BMI > 30 and row.age > 40 and row.Blood_cholesterol < 159:
        z3 = np.random.randint(159,200) + np.random.rand()
        df.at[index, 'Blood_cholesterol'] = z3
        newBloodChol.append(z3)
    else:
        z3 = df.at[index, 'Blood_cholesterol']
        newBloodChol.append(z3) 
        
# Just checking the lenght of the array
len(newBloodChol)
```

And the final table would be:

```python
df = pd.DataFrame({'age': age, 'gender': gender, 'Family history': fHist, 'Blood_cholesterol': newBloodChol, 'Tobacco consumption': smokers, 'BMI': BMI})
# to start the index from 1 rather than 0
df.index = np.arange(1, len(df) + 1)
# Align the columns at the center with the style feature
df.style.set_properties(**{'text-align': 'center'})
```

Then we can just relate all the variables together to identify relationships between them or spot trends or isolated facts.

```python
finalDS = pd.DataFrame({'age': age, 'gender': gender, 'Family history': fHist, 'Blood cholesterol': bloodChol, 'Tobacco consumption': smokers, 'BMI': BMI})
# with height we can set the size of the plots, the higher the number, the bigger the plots 
sns.pairplot(finalDS, height=5)
plt.show()
```

The histograms are irrelevant because they just show the same variable on both axis. When plotting blood cholesterol and BMI together we see that all the dots are well spread, same as what happens between age and BMI. However age and blood cholesterol shows something different, it looks like two steps of a ladder, there are two ranges, and this has to do with the modification of the blood cholesterol depending on the age. 

On the other hand I synthetised six variables, being values three of them and strings the other three. With the `hue` parameter we can colour the dots per groups. As an example I'll choose the gender hue:

```python
finalDS = pd.DataFrame({'age': age, 'gender': gender, 'Family history': fHist, 'Blood cholesterol': bloodChol, 'Tobacco consumption': smokers, 'BMI': BMI})
# with height we can set the size of the plots, the higher the number, the bigger the plots 
sns.pairplot(finalDS, hue='gender', height=5)
plt.show()
```

And this way we can clearly spot the difference of cholesterol level between man and woman between the ages 45 and 55, as I set when synthetising that array.
